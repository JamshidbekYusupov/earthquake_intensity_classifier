```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
url = 'https://www.weather.go.kr/neng/earthquake/earthquake-korea.do?schOption=&xls=0&startTm=2026-01-01&endTm=2026-02-16&startSize=&endSize=&startLat=&endLat=&startLon=&endLon=&lat=&lon=&dist=&keyword=&dpType=a'
```

```python
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026]
```

```python
rows = []

for year in years:

    print(year)

    url = f'https://www.weather.go.kr/neng/earthquake/earthquake-korea.do?schOption=&xls=0&startTm={year}-01-01&endTm=2026-02-16&startSize=&endSize=&startLat=&endLat=&startLon=&endLon=&lat=&lon=&dist=&keyword=&dpType=a'

    # User agent qurish / Building user agent 
    headers = {
        'User-Agent': 'Mozilla/5.0'
    }

    response = requests.get(url, headers=headers)
    response

    soup = BeautifulSoup(response.content,  'html.parser')

    table = soup.find('table', {"class": "table-col eqk-search-table"})

    headers =  [th.get_text(strip = True) for th in table.find_all('th')]

    for row in table.find('tbody').find_all('tr'):
        cols = [td.get_text(strip = True) for td in row.find_all('td')]
        if cols:
            rows.append(cols)
    
    df = pd.DataFrame(rows, columns=headers[:len(rows[0])])
```

```python
df.info()
```

```python
df.nunique()
```

```python
out_path = r'C:\project_5\Data\Scraped_Data'

os.makedirs(out_path, exist_ok=True)

data_path = os.path.join(out_path, 'df_1.csv')

df.to_csv(data_path, index=False)
```
